## Download Base models (Legacy)

* Not recommended, you may use Batchlink Downloader extension on SD (integrated) later

```python
#Base models:

model_link1 = ''
model_link2 = ''
model_link3 = ''
model_link4 = ''
model_link5 = ''














#Run this to download base models:

import os
import requests
from tqdm import tqdm

# Function to download a file with progress
def download_file(url, target_dir):
    filename = url.split("/")[-1]  # Extract the filename from the URL
    download_path = os.path.join(target_dir, filename)
    response = requests.get(url, stream=True)
    total_size = int(response.headers.get("content-length", 0))
    block_size = 1024  # 1 Kibibyte

    with open(download_path, "wb") as file, tqdm(
        desc=filename,
        total=total_size,
        unit="B",
        unit_scale=True,
        unit_divisor=1024,
    ) as bar:
        for data in response.iter_content(block_size):
            file.write(data)
            bar.update(len(data))



#Base models
download_links = [model_link1,
                  model_link2,
                  model_link3,
                  model_link4,
                  model_link5
                 ]


target_dir = "/home/studio-lab-user/content/stable-diffusion-webui/models/Stable-diffusion/"

# Loop through the download links and download files to the target directory
for link in download_links:
    if link == '':
        pass
    else:
        download_file(link, target_dir)
```
